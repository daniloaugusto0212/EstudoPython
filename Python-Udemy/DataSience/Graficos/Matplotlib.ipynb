```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40], [1, 10, 7, 8], label='Sala 1', color='red', linestyle='dotted')
plt.plot([10, 22, 35, 45], [1, 15, 9, 12], label='Sala 2', color='black', linestyle='dashed')
plt.ylabel('alunos')
plt.xlabel('notas')
plt.title('Relação entre ID do aluno e notas')
plt.legend()
plt.show()
```
